```python
#importing core libraries
import pandas as pd #pip install pandas
import numpy as np #pip install numpy

melbourne_file_path = 'C:\\Users\\knguy\\GitHub\\Elevvo_InternProjects\\Introduction\\melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path)
print(melbourne_data.columns)
```

```python
# Drops missing values (think of na as "not available")
melbourne_data = melbourne_data.dropna(axis=0)
print(melbourne_data.columns)

# Prints out the first few entries from the Price column
y = melbourne_data.Price
print(y.head())

# Prints out list of chosen features
malbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']   
X = melbourne_data[malbourne_features]
print(X.head())

# Prints out a summary of the data in the chosen features
X.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor #pip install scikit-learn

# Defines the model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fits the model
melbourne_model.fit(X, y)
```

```python
# Makes predictions
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```
